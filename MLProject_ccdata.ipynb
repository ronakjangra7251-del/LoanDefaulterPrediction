```python
# perform label encoding and uodate all the object columns

from sklearn.preprocessing import LabelEncoder
encoding_maps = {}
le = LabelEncoder()

for col in df.select_dtypes('object').columns:
  df[col] = le.fit_transform(df[col])

  encoding_maps[col] = dict(zip(le.classes_, le.transform(le.classes_)))

# Now you can see the mappings
for col, mapping in encoding_maps.items():
    print(f"{col} mapping: {mapping}")
```

```python
df.select_dtypes('number')
```

```python
df.drop(['SK_ID_CURR'], axis=1, inplace=True)
```

```python
df.describe()
```

```python
# calculate the percentage of null value in each column
filtered_columns = []
for col in df.columns:
  if df[col].isnull().sum()/len(df)*100<40:
    filtered_columns.append(col)
```

```python
df = df[filtered_columns]
```

```python
df.isnull().sum()
```

```python
df["OBS_30_CNT_SOCIAL_CIRCLE"].value_counts()
```

```python
df.drop(columns="EXT_SOURCE_3",inplace=True)
```

```python
newdf = df.dropna()
newdf.shape
```

```python
input = newdf.drop(['TARGET'],axis = 1)
output = newdf["TARGET"]
```

```python
# perform feature selection using SelectKBest technique and select 50 important methods

from sklearn.feature_selection import SelectKBest, f_classif

fs = SelectKBest(f_classif, k=20)
fs.fit(input, output)
```

```python
#pd.DataFrame({"Columns":input.columns,"Scores":fs.scores_}).sort_values(by="Scores",ascending=False)
```

```python
input2 = input[fs.get_feature_names_out()]
```

```python
# lets select best 30 features using wrapper method from above method

from sklearn.feature_selection import RFE
from sklearn.ensemble import RandomForestClassifier

rfc = RandomForestClassifier(n_estimators=100,max_depth=6,n_jobs=-1,random_state=42,max_features="sqrt")
rfe = RFE(rfc,n_features_to_select=10,verbose=True)
rfe.fit(input2,output)
```

```python
final_input = input2[rfe.get_feature_names_out()]
final_input
```

#lets balance data with SMOTE sampling on a part of data

```python
finaldf = pd.concat([final_input,output],axis = 1)
finaldf.shape
```

```python
sample_df = finaldf.sample(frac = 0.20,random_state=42)
sample_df.shape
X = sample_df.drop("TARGET",axis=1)
y = sample_df["TARGET"]
```

```python
import pandas as pd
from imblearn.over_sampling import SMOTE

# Apply SMOTE
smote = SMOTE(random_state=42)
X_res, y_res = smote.fit_resample(X, y)

# Convert back to DataFrame (optional)
X_resampled = pd.DataFrame(X_res, columns=X.columns)
y_resampled = pd.Series(y_res, name="TARGET")

# Final balanced dataset
balanced_df = pd.concat([X_resampled, y_resampled], axis=1)
```

```python
balanced_df.shape
```

```python
X = balanced_df.drop("TARGET",axis = 1)
y = balanced_df["TARGET"]
```

```python
#lets split data using train_test_split
from sklearn.model_selection import train_test_split

X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=4)
```

```python
#lets train some models on balanced data
# lets import all models that we are going to train
from sklearn.metrics import accuracy_score, f1_score, roc_auc_score
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.svm import SVC
import xgboost as xgb
```

```python
model = {"RandomForest": RandomForestClassifier(n_estimators=20,max_depth=10,n_jobs=-1,random_state=42),
         "XGBoost": xgb.XGBClassifier(n_estimators=20,tree_method="hist",eval_metric = "logloss",random_state = 42),
         "Gradient Boosting": GradientBoostingClassifier(n_estimators=50, random_state=42)}

results = []
scores = []
```

```python
import joblib
for name, model in model.items():
    model.fit(X_train, y_train)
    train_score = model.score(X_train, y_train)  # Accuracy on training set
    test_score = model.score(X_test, y_test)    # Accuracy on test set

    results.append({"Model": name, "Train Score": train_score, "Test Score": test_score})

    joblib.dump(model, f"{name}.pkl")
    print(f"{name} trained and saved as {name}.pkl")

# Convert to DataFrame for display
results_df = pd.DataFrame(results).set_index("Model")
print(results_df)
```
